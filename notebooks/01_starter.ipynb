```python
import pandas as pd
from pathlib import Path
import glob as gb
import numpy as np
```

```python
data_dir = gb.glob("../data/raw/*.csv")

datasets = {}
for file_path in data_dir:
    datasets[Path(file_path).name] = pd.read_csv(file_path)

print(f"Loaded {len(datasets)} files")
    
```

```python
for key,value in datasets.items():
    print("Data set file", key)
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    print(df[' Label'].value_counts())
    print(df[' Label'].value_counts(normalize=True))
    print("Unique Label Type",df[' Label'].unique())
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            print(f"Column '{col}' has {df[col].isnull().sum()} missing values.")
        numeric_col = pd.to_numeric(df[col], errors='coerce') # Convert the column to numeric, coercing errors to NaN (need more explanation)
        if np.isposinf(numeric_col).sum() > 0:
            print(f"Column '{col}' contains {np.isposinf(numeric_col).sum()} positive infinite values.")
        if np.isneginf(numeric_col).sum() > 0:
            print(f"Column '{col}' contains negative infinite values.")
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    print(f"Number of duplicate rows: {df.duplicated().sum()}")
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.replace(' ', '_')
    df['Label'] = df['Label'].str.replace('�', '_')
    df['Label'] = df['Label'].str.replace(' ', '_')
    print(df.columns)
    print(df['Label'].unique())
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    df.drop_duplicates(inplace=True)
    df.replace(np.inf, np.nan, inplace=True)
    numeric_cola = df.select_dtypes(include=np.number).columns # explain this line
    df[numeric_cola] = df[numeric_cola].fillna(df[numeric_cola].median()) # explain this line
    print("Data set file", key)
    print(f"Number of duplicate rows after dropping duplicates: {df.duplicated().sum()}")
    print(f"Number of missing values after filling: {df.isnull().sum().sum()}")
    print(f"Number of infinite values after replacing: {np.isinf(df.select_dtypes(include=[np.number])).sum().sum()}") # what df.select_dtypes(include=[np.number]) returns
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value 
    df['isAttack'] = (df['Label'] != 'BENIGN').astype(int) # anything that is not BENIGN will be considered as attack, and we will assign it a value of 1, while BENIGN will be assigned a value of 0
    print("Data set file", key)
    print(df['isAttack'].value_counts())
    df.drop(columns=['Label'], inplace=True)
```

```python
df_combined = pd.concat(datasets.values(), ignore_index=True)
df_combined.to_csv("../data/processed/combined_dataset.csv", index=False)
```

```python
combined_data = pd.read_csv("../data/processed/combined_dataset.csv")
```

```python
print(len(combined_data))
print(combined_data.shape)
print(combined_data.isnull().sum().sum())
print(combined_data['isAttack'].value_counts())
print("Duplicate rows:", combined_data.duplicated().sum())
print(combined_data.columns)    
```

```python
combined_data.drop_duplicates(inplace=True)
combined_data.duplicated().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
```

```python
target_y = combined_data['isAttack']
features_X = combined_data.drop(columns=['isAttack'])
x_train, x_test, y_train, y_test = train_test_split(features_X, target_y, test_size=0.2, random_state=42)
model = RandomForestClassifier(random_state=42, n_estimators=100)
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
acc = accuracy_score(y_test, prediction)
print(f"Accuracy of the model: {acc:.4f}")
con_matrix = confusion_matrix(y_test, prediction)
print("Confusion Matrix:")
print(con_matrix)
pre = precision_score(y_test, prediction)
print(f"Precision: {pre:.4f}")
rec = recall_score(y_test, prediction)
print(f"Recall: {rec:.4f}")
f_score = f1_score(y_test, prediction)
print(f"F1 Score: {f_score:.4f}")
C_report = classification_report(y_test, prediction)
print("Classification Report:")
print(C_report)
```

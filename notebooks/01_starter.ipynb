```python
import pandas as pd
from pathlib import Path
import glob as gb
import numpy as np
```

```python
data_dir = gb.glob("../data/raw/*.csv")

datasets = {}
for file_path in data_dir:
    datasets[Path(file_path).name] = pd.read_csv(file_path)

print(f"Loaded {len(datasets)} files")
    
```

```python
for key,value in datasets.items():
    print("Data set file", key)
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    print(df[' Label'].value_counts())
    print(df[' Label'].value_counts(normalize=True))
    print("Unique Label Type",df[' Label'].unique())
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            print(f"Column '{col}' has {df[col].isnull().sum()} missing values.")
        numeric_col = pd.to_numeric(df[col], errors='coerce') # Convert the column to numeric, coercing errors to NaN (need more explanation)
        if np.isposinf(numeric_col).sum() > 0:
            print(f"Column '{col}' contains {np.isposinf(numeric_col).sum()} positive infinite values.")
        if np.isneginf(numeric_col).sum() > 0:
            print(f"Column '{col}' contains negative infinite values.")
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    print(f"Number of duplicate rows: {df.duplicated().sum()}")
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    print("Data set file", key)
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.replace(' ', '_')
    df['Label'] = df['Label'].str.replace('�', '_')
    df['Label'] = df['Label'].str.replace(' ', '_')
    print(df.columns)
    print(df['Label'].unique())
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value
    df.drop_duplicates(inplace=True)
    df.replace(np.inf, np.nan, inplace=True)
    numeric_cola = df.select_dtypes(include=np.number).columns # explain this line
    df[numeric_cola] = df[numeric_cola].fillna(df[numeric_cola].median()) # explain this line
    print("Data set file", key)
    print(f"Number of duplicate rows after dropping duplicates: {df.duplicated().sum()}")
    print(f"Number of missing values after filling: {df.isnull().sum().sum()}")
    print(f"Number of infinite values after replacing: {np.isinf(df.select_dtypes(include=[np.number])).sum().sum()}") # what df.select_dtypes(include=[np.number]) returns
    print("====================================================================")
```

```python
for key,value in datasets.items():
    df = value 
    df['isAttack'] = (df['Label'] != 'BENIGN').astype(int) # anything that is not BENIGN will be considered as attack, and we will assign it a value of 1, while BENIGN will be assigned a value of 0
    print("Data set file", key)
    print(df['isAttack'].value_counts())
    df.drop(columns=['Label'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import BorderlineSMOTE
```

```python
Mon_data = datasets['Monday-WorkingHours.pcap_ISCX.csv']
tues_data = datasets['Tuesday-WorkingHours.pcap_ISCX.csv']
wed_data = datasets['Wednesday-workingHours.pcap_ISCX.csv']
train_data = pd.concat([Mon_data, tues_data, wed_data], ignore_index=True)
thurs_data = datasets['Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv']
thurs_data2 = datasets['Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv']
valid_data = pd.concat([thurs_data, thurs_data2], ignore_index=True)
friday_data = datasets['Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv']
friday_data2 = datasets['Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv']
friday_data3 = datasets['Friday-WorkingHours-Morning.pcap_ISCX.csv']
test_data = pd.concat([friday_data, friday_data2,friday_data3], ignore_index=True)
print(f"Training data shape: {train_data.shape}")
print(f"Validation data shape: {valid_data.shape}")
print(f"Test data shape: {test_data.shape}")
```

```python
# y_train = train_data['isAttack']
# x_train = train_data.drop(columns=['isAttack'])
model = RandomForestClassifier(class_weight='balanced',random_state=42, n_estimators=100)
# model.fit(x_train, y_train)
```

```python
y_valid_data = valid_data['isAttack']
x_valid_data = valid_data.drop(columns=['isAttack'])
y_valid_pred = model.predict(x_valid_data)
```

```python
y_train = train_data['isAttack']
x_train = train_data.drop(columns=['isAttack'])
sm = SMOTE(random_state=42) # explain this line
x_train_resampled, y_train_resampled = sm.fit_resample(x_train, y_train)
model.fit(x_train_resampled, y_train_resampled)
```

```python
y_valid_data = valid_data['isAttack']
x_valid_data = valid_data.drop(columns=['isAttack'])
y_valid_pred = model.predict_proba(x_valid_data)[:, 1]
chosen_threshold = 0.5
preds = (y_valid_pred >= chosen_threshold).astype(int)
accuracy = accuracy_score(y_valid_data, preds)
print(f"Validation Accuracy: {accuracy:.4f}")
f1 = f1_score(y_valid_data, preds)
print(f"Validation F1 Score: {f1:.4f}")
conf_matr = confusion_matrix(y_valid_data, preds)
print("Confusion Matrix:", conf_matr)
per_score = precision_score(y_valid_data, preds)
print(f"Validation Precision Score: {per_score:.4f}")
rec_score = recall_score(y_valid_data, preds)
print(f"Validation Recall Score: {rec_score:.4f}")
class_report = classification_report(y_valid_data, preds)
print("Classification Report:\n", class_report)
```

```python
y_valid_data = valid_data['isAttack']
x_valid_data = valid_data.drop(columns=['isAttack'])
y_valid_pred = model.predict_proba(x_valid_data)[:, 1]
chosen_threshold = [0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.7]
for i in chosen_threshold:
    print(f"Threshold: {i}")
    preds = (y_valid_pred >= i).astype(int)
    accuracy = accuracy_score(y_valid_data, preds)
    print(f"Validation Accuracy: {accuracy:.4f}")
    f1 = f1_score(y_valid_data, preds)
    print(f"Validation F1 Score: {f1:.4f}")
    conf_matr = confusion_matrix(y_valid_data, preds)
    print("Confusion Matrix:", conf_matr)
    per_score = precision_score(y_valid_data, preds)
    print(f"Validation Precision Score: {per_score:.4f}")
    rec_score = recall_score(y_valid_data, preds)
    print(f"Validation Recall Score: {rec_score:.4f}")
    class_report = classification_report(y_valid_data, preds)
    print("Classification Report:\n", class_report)
    print("====================================================================")
```

```python
y_test_data = test_data['isAttack']
x_test_data = test_data.drop(columns=['isAttack'])
y_test_pred = model.predict_proba(x_test_data)[:, 1]
preds = (y_test_pred >= 0.15).astype(int)
accuracy = accuracy_score(y_test_data, preds)
print(f"Test Accuracy: {accuracy:.4f}")
f1 = f1_score(y_test_data, preds)
print(f"Test F1 Score: {f1:.4f}")
conf_matr = confusion_matrix(y_test_data, preds)
print("Confusion Matrix:", conf_matr)
per_score = precision_score(y_test_data, preds)
print(f"Test Precision Score: {per_score:.4f}")
rec_score = recall_score(y_test_data, preds)
print(f"Test Recall Score: {rec_score:.4f}")
class_report = classification_report(y_test_data, preds)
print("Classification Report:\n", class_report)
```
